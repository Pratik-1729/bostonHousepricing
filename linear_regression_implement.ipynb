```python
X=boston_df.iloc[:,:-1]
y = boston_df.iloc[:,-1]
```

```python
X.head()
```

```python
#train and test
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,root_mean_squared_error,accuracy_score
from sklearn.preprocessing import StandardScaler
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
X_train
```

```python
scale = StandardScaler()
X_train = scale.fit_transform(X_train)
X_test = scale.transform(X_test)
```

```python
import pickle
pickle.dump(scale,open('scaling.pkl','wb'))
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
reg.fit(X_train,y_train)
```

```python
print(reg.coef_)
```

```python
print(reg.intercept_)
```

```python
reg_pred = reg.predict(X_test)
```

```python
plt.scatter(y_test,reg_pred)
```

```python
residual = y_test - reg_pred
```

```python
residual
```

```python
sns.displot(residual,kind='kde')
```

```python
print(root_mean_squared_error(y_test,reg_pred))
```

```python
print(r2_score(y_test,reg_pred))
```

```python
sample = np.array([0.1, 18.0, 2.3, 0, 0.5, 6.0, 65.0, 4.0, 1, 300.0, 15.0, 500.0, 5.0])
sample = sample.reshape(1, -1)  # Reshape to (1, 13)

prediction = reg.predict(scale.transform(sample))
print(f"Predicted price: {prediction[0]:.2f}")

```
